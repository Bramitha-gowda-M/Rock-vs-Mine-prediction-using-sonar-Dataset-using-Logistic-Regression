Train_Test_split

```python
from sklearn.model_selection import train_test_split
```

```python
X = data.drop("Freq_61",axis = 1)
```

```python
y = data["Freq_61"]
```

```python
X.shape
```

```python
X
```

```python
y.shape
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.1,random_state=1)
```

```python
X_train.shape
```

```python
X_test.shape
```

Trining model

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
preditct = model.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
ac = accuracy_score(preditct, y_train)
```

```python
print(ac)
```

```python
preditct = model.predict(X_test)
act = accuracy_score(preditct, y_test)
```
